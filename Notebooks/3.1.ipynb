# üß† Predicci√≥n de Riesgo de Cr√©dito ‚Äì German Credit Dataset

## üìå Contexto del problema

Cuando un banco recibe una solicitud de pr√©stamo, debe decidir si aprueba o no el cr√©dito con base en el perfil del solicitante. Existen dos tipos de riesgo:

- Si el solicitante es un **buen riesgo crediticio** (es decir, probablemente pagar√°), y se **rechaza** su solicitud, el banco **pierde una oportunidad de negocio**.
- Si el solicitante es un **mal riesgo crediticio** (es decir, probablemente no pagar√°), y se **aprueba** su solicitud, el banco **asume una p√©rdida financiera**.

Este dataset incluye 1000 registros con 20 variables predictoras, adem√°s de la clasificaci√≥n `Creditability` (bueno o malo), que es lo que intentaremos predecir.

---
## üï∞Ô∏è Consideraciones hist√≥ricas y monetarias

Este dataset fue recolectado antes de la adopci√≥n del euro (‚Ç¨) como moneda oficial en Alemania. Por lo tanto:

- Todas las cantidades monetarias (como `Credit_Amount`, `Savings/Stocks`, etc.) est√°n expresadas en **marcos alemanes (Deutsche Mark, DM)**.
- Alemania cambi√≥ al euro en el a√±o **2002**, y el tipo de cambio final fue:
  > **1 EUR = 1.95583 DM**

Aunque no es necesario convertir las cantidades a euros para el an√°lisis (ya que se trabajar√° con modelos relativos), **es importante tener presente este contexto**:

- Las cifras no representan montos actuales.
- Los rangos de ingreso, empleo y ahorro deben interpretarse en relaci√≥n al est√°ndar de vida en Alemania antes del 2002.
- Este detalle puede influir en la interpretaci√≥n del poder adquisitivo y la clasificaci√≥n del riesgo crediticio en su momento.

üîé *Conclusi√≥n: trabajaremos con las cifras tal como est√°n, pero entendiendo que no corresponden al valor monetario moderno.*

### üìä Relaci√≥n entre variables num√©ricas y `Creditability`

Se exploraron tres variables num√©ricas clave en funci√≥n del riesgo crediticio (`Creditability`):

#### üü¶ 1. `Credit_Amount`
- Se observa que los solicitantes con menor `Creditability` (riesgo alto, valor 0) tienden a solicitar montos de cr√©dito ligeramente mayores.
- Hay una mayor concentraci√≥n de outliers (montos extremos) en esta clase, lo cual podr√≠a estar asociado al incumplimiento.

#### üü¶ 2. `Age_years`
- La edad media de los solicitantes no muestra una diferencia muy marcada entre clases.
- Sin embargo, la dispersi√≥n parece ligeramente mayor entre los buenos pagadores, lo que sugiere que hay perfiles de riesgo en m√∫ltiples grupos de edad.

#### üü¶ 3. `Credit_Duration_Months`
- Los malos pagadores (`Creditability` = 0) tienden a tener una duraci√≥n de cr√©dito m√°s prolongada.
- Esta variable podr√≠a ser predictiva, ya que un mayor plazo puede implicar mayor riesgo de incumplimiento.

üìå En resumen, variables como **`Credit_Amount` y `Credit_Duration_Months`** muestran una mayor diferenciaci√≥n seg√∫n el riesgo, y podr√≠an ser relevantes para el modelo.


```python
from scipy.stats import ttest_ind, chi2_contingency

variables = ['Credit_Amount', 'Age_years', 'Duration_of_Credit_monthly']

for var in variables:
    group1 = df[df['Creditability'] == 1][var]
    group0 = df[df['Creditability'] == 0][var]
    t_stat, p_val = ttest_ind(group1, group0, equal_var=False)
    print(f"{var}: t = {t_stat:.3f}, p = {p_val:.4f}")
```

### üß™ Prueba t de Student ‚Äì Comparaci√≥n de medias entre grupos (`Creditability`)

Se aplic√≥ una prueba t para muestras independientes (Welch's t-test) para comparar las medias de variables num√©ricas clave entre:

- Grupo 1: solicitantes considerados buen riesgo (`Creditability = 1`)
- Grupo 0: solicitantes considerados mal riesgo (`Creditability = 0`)

#### üìå Resultados:

| Variable                     | t-statistic | p-valor  | Interpretaci√≥n                                  |
|-----------------------------|-------------|----------|------------------------------------------------|
| `Credit_Amount`             | -4.264      | 0.0000   | Diferencia significativa: los malos pagadores tienden a solicitar montos mayores. |
| `Age_years`                 | 2.908       | 0.0038   | Diferencia significativa: los buenos pagadores son, en promedio, un poco mayores. |
| `Duration_of_Credit_monthly`| -6.470      | 0.0000   | Diferencia significativa: los malos pagadores tienen cr√©ditos m√°s largos. |

‚úÖ En todos los casos, el p-valor < 0.05, por lo tanto **rechazamos la hip√≥tesis nula**: existe evidencia estad√≠stica de que las medias difieren entre los dos grupos.

Estas variables pueden ser relevantes para predecir el riesgo crediticio y deber√≠an considerarse en la fase de modelado.


```python
categorical_cols = [
    'Account_Balance',
    'Payment_Status_of_Previous_Credit',
    'Purpose',
    'Value_Savings_Stocks',
    'Length_of_current_employment',
    'Instalment_per_cent',
    'Sex_Marital_Status',
    'Guarantors',
    'Most_valuable_available_asset',
    'Concurrent_Credits',
    'Type_of_apartment',
    'No_of_Credits_at_this_Bank',
    'Occupation',
    'No_of_dependents',
    'Duration_in_Current_address',
    'Telephone',
    'Foreign_Worker'
]

# Aplicar prueba chi-cuadrado para cada una
for col in categorical_cols:
    table = pd.crosstab(df[col], df['Creditability'])
    chi2, p, dof, expected = chi2_contingency(table)
    print(f"{col}: chi2 = {chi2:.2f}, p = {p:.4f}")
```
